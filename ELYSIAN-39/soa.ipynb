```python
import pandas as pd
```

```python
import os
ls = os.listdir("acc")
```

```python
df = pd.read_csv('acc/'+ls[1])
```

```python
for i in ls:
    t_df = pd.read_csv('acc/'+i)
    df = pd.concat([df,t_df],axis=0)
```

```python
df.reset_index()
```

```python
df.drop(df.columns[[3,7,8,9,10,11,12,13,14,15]],axis=1)
```

```python
df.to_csv('soa.csv')
```

```python
from PyPDF2 import PdfWriter, PdfReader
pages_to_delete = [3, 4, 5] # page numbering starts from 0
infile = PdfReader('acc.pdf', 'rb')
output = PdfWriter()

for i in range(len(infile.pages) ):
    if i not in pages_to_delete:
        p = infile.get_page(i)
        output.add_page(p)

with open('newfile.pdf', 'wb') as f:
    output.write(f)
```
